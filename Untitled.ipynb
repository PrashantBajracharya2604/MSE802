```python
import numpy as np

# 1. Define your 1-bit function f(x) where x can be 0 or 1
def f(x):
    return 1 if x == 1 else 0

# 2. Setup the basis states in order: |00>, |01>, |10>, |11>
# In standard big-endian format: |xy> 
# |00> is index 0, |01> is index 1, |10> is index 2, |11> is index 3
basis_states = [(0,0), (0,1), (1,0), (1,1)]
matrix_size = 4
oracle_matrix = np.zeros((matrix_size, matrix_size), dtype=int)

# 3. Fill the matrix by evaluating the mapping for each column
for col_idx, (x, y) in enumerate(basis_states):
    # Compute the transformation: y_new = y ⊕ f(x)
    y_new = y ^ f(x) 
    x_new = x  # x remains unchanged
    
    # Find which row index this new state corresponds to
    row_idx = basis_states.index((x_new, y_new))
    
    # Place a 1 in the matrix to map: col_idx -> row_idx
    oracle_matrix[row_idx, col_idx] = 1

print("Generated 4x4 Oracle Matrix for |x, y> -> |x, y ⊕ f(x)>:\n")
print(oracle_matrix)
```
