# Understanding
## The Goal: Constant vs. Balanced Functions

The objective is to determine whether a hidden "black box" function (an oracle) is **constant** or **balanced** using the absolute minimum number of queries.

* **Constant:** The function always returns the exact same output for any input.
* **Balanced:** The function returns `0` for exactly half of the possible inputs and `1` for the other half.

---

## Classical vs. Quantum Complexity

### The Classical Approach (Deterministic)
To prove with 100% certainty what the function does, a standard computer must check more than half of all possible input combinations in a worst-case scenario.
* For a system with **n** inputs, there are **2^n** total combinations.
* Queries needed **2^(n-1) + 1** .
* *Example:* If n = 10 (1,024 combinations), number of time the box tests **513 times**. As n grows, this number also grows exponentially.

### The Quantum Approach (Deutsch-Jozsa Algorithm)
By mapping into a quantum permutation matrix (U_f), a quantum computer evaluates the structural symmetry of the box.
* It determines if the function is constant or balanced in **exactly 1 single query**.
* This remains true for any value of n.

```python
# Activity 1 ; this script builds the unitary matrix U_f for the Deutsch-Jozsa algorithm based on function mapping provided by the user.
# generated with AI
import numpy as np

def main():
    print("=======================================================")
    print("   DEUTSCH-JOZSA MATRIX GENERATOR FROM CUSTOM MAPPING  ")
    print("=======================================================")
    
    # 1. Ask the user for the number of input qubits (n)
    try:
        n = int(input("Enter the number of input qubits (n): ").strip())
        if n < 1:
            print("n must be at least 1. Defaulting to n=2.")
            n = 2
    except ValueError:
        print("Invalid input format. Defaulting to n=2.")
        n = 2

    num_inputs = 2 ** n
    mapping_dict = {}
    
    print(f"\nEnter the output (0 or 1) for each of the {num_inputs} input states:")
    print("-" * 55)
    
    # 2. Dynamically loop through all binary states based on the chosen n
    for i in range(num_inputs):
        # Generate the bitstring representation of length n
        bitstring = format(i, f'0{n}b')
        
        # Direct validation loop to ensure the user only enters a 0 or 1
        while True:
            val = input(f"State |{bitstring}⟩ maps to (0 or 1): ").strip()
            if val in ['0', '1']:
                mapping_dict[bitstring] = int(val)
                break
            print("❌ Invalid entry. You must enter exactly 0 or 1.")

    # 3. Calculate the dimensions and build the oracle matrix
    matrix_dim = 2 ** (n + 1)
    U_f = np.zeros((matrix_dim, matrix_dim), dtype=int)
    
    # Map configurations: |x, y⟩ -> |x, y ⊕ f(x)⟩
    for bitstring, fx in mapping_dict.items():
        x_int = int(bitstring, 2)
        for y in [0, 1]:
            new_y = y ^ fx
            row_index = (x_int << 1) + y
            col_index = (x_int << 1) + new_y
            U_f[col_index, row_index] = 1

    # 4. Display the finalized matrix layout
    print("\n" + "=" * 60)
    print("                     GENERATED MATRIX                  ")
    print("=" * 60)
    print(f"• Input Qubits (n) : {n}")
    print(f"• Total Qubits     : {n + 1} (Includes 1 Ancilla Qubit)")
    print(f"• Matrix Dimensions: {matrix_dim}x{matrix_dim}")
    print(f"• Your Mapping     : {mapping_dict}\n")
    
    print(f"THE FULL {matrix_dim}x{matrix_dim} ORACLE MATRIX (U_f):")
    print("-" * (matrix_dim * 3 + 4))
    for row in U_f:
        print("  " + "  ".join(str(val) for val in row))
    print("-" * (matrix_dim * 3 + 4))
    print("=" * 60)

if __name__ == "__main__":
    main()
```

```python
# Activity 1 ; this script builds the unitary matrix U_f for the Deutsch-Jozsa algorithm based on function mapping provided by the user. and the circuit.
import cirq
import numpy as np
from collections import Counter

# Define a custom gate so Cirq prints a clean "U_f" block instead of raw matrix text
class OracleGate(cirq.Gate):
    def __init__(self, matrix):
        super().__init__()
        self.matrix = matrix

    def _num_qubits_(self) -> int:
        return int(np.log2(self.matrix.shape[0]))

    def _unitary_(self):
        return self.matrix

    def _circuit_diagram_info_(self, args) -> cirq.CircuitDiagramInfo:
        # Labels every qubit wire inside this multi-qubit block as 'U_f'
        return cirq.CircuitDiagramInfo(wire_symbols=("U_f",) * self._num_qubits_())

def main():
    print("=======================================================")
    print("  DEUTSCH-JOZSA INTERACTIVE CIRCUIT & MATRIX GENERATOR ")
    print("=======================================================")
    
    # 1. Ask the user for the number of input qubits (n)
    try:
        n = int(input("Enter the number of input qubits (n): ").strip())
        if n < 1:
            print("n must be at least 1. Defaulting to n=2.")
            n = 2
    except ValueError:
        print("Invalid input format. Defaulting to n=2.")
        n = 2

    num_inputs = 2 ** n
    mapping_dict = {}
    
    print(f"\nEnter the output (0 or 1) for each of the {num_inputs} input states:")
    print("-" * 55)
    
    # 2. Dynamically loop through all binary states based on the chosen n
    for i in range(num_inputs):
        bitstring = format(i, f'0{n}b')
        
        while True:
            val = input(f"State |{bitstring}⟩ maps to (0 or 1): ").strip()
            if val in ['0', '1']:
                mapping_dict[bitstring] = int(val)
                break
            print("❌ Invalid entry. You must enter exactly 0 or 1.")

    # 3. Calculate the dimensions and build the oracle matrix
    matrix_dim = 2 ** (n + 1)
    U_f = np.zeros((matrix_dim, matrix_dim), dtype=int)
    
    # Map configurations: |x, y⟩ -> |x, y ⊕ f(x)⟩
    for bitstring, fx in mapping_dict.items():
        x_int = int(bitstring, 2)
        for y in [0, 1]:
            new_y = y ^ fx
            row_index = (x_int << 1) + y
            col_index = (x_int << 1) + new_y
            U_f[col_index, row_index] = 1

    # 4. Build Cirq Circuit dynamically based on user setup
    qubits = cirq.LineQubit.range(n + 1)
    inputs = qubits[:n]
    ancilla = qubits[n]

    circuit = cirq.Circuit()
    circuit.append(cirq.X(ancilla))                                # Flip ancilla line to |1⟩
    circuit.append([cirq.H(q) for q in qubits])                    # Put everything in superposition
    
    # Inject the matrix wrapped inside our custom visual gate
    circuit.append(OracleGate(U_f.astype(float))(*qubits))  
    
    circuit.append([cirq.H(q) for q in inputs])                     # Apply final Hadamards to inputs
    circuit.append([cirq.measure(q, key=f'result_{i}') for i, q in enumerate(inputs)]) # Measure inputs

    # 5. Simulate the quantum execution
    simulator = cirq.Simulator()
    result = simulator.run(circuit, repetitions=100)

    # Clean measurement string parsing
    outcomes = []
    for shot in range(100):
        shot_bits = []
        for idx in range(n):
            measurement_key = f'result_{idx}'
            bit_value = int(result.measurements[measurement_key][shot][0])
            shot_bits.append(str(bit_value))
        bitstring_outcome = "".join(shot_bits)
        outcomes.append(bitstring_outcome)

    # 6. Display the finalized outputs
    print("\n" + "=" * 60)
    print("                     SIMULATION RESULTS                ")
    print("=" * 60)
    print(f"• Input Qubits (n) : {n}")
    print(f"• Total Qubits     : {n + 1} (Includes 1 Ancilla Qubit)")
    print(f"• Matrix Dimensions: {matrix_dim}x{matrix_dim}")
    print(f"• Your Mapping     : {mapping_dict}\n")
    
    print("THE GENERATED QUANTUM CIRCUIT DIAGRAM:")
    print("-" * 60)
    print(circuit)
    print("-" * 60)
    
    print(f"\nTHE FULL {matrix_dim}x{matrix_dim} ORACLE MATRIX (U_f):")
    print("-" * (matrix_dim * 3 + 4))
    for row in U_f:
        print("  " + "  ".join(str(val) for val in row))
    print("-" * (matrix_dim * 3 + 4))
    
    print(f"\nQUANTUM SIMULATION OUTCOMES (100 shots):")
    print(f"👉 {dict(Counter(outcomes))}")
    print("=" * 60)

if __name__ == "__main__":
    main()
```
