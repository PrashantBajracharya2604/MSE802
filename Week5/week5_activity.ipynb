**Task 1 : Perform all operations on matrix.**

```python
import numpy as np

# Complex matrices
A = np.array([[1+2j, 3-1j],
              [2+1j, 4-2j]])

B = np.array([[3-1j, 2+2j],
              [1-2j, 2-1j]])


print("Matrix A:\n", A)
print("\nMatrix B:\n", B)

# 1. Transpose
print("\n1. Transpose of A:\n", np.transpose(A))

# 2. Conjugate Transpose / Adjoint
A_H = np.conj(A.T)
print("\n2. Conjugate Transpose (Adjoint) of A:\n", A_H)

# 3. Matrix Multiplication
print("\n3. Matrix Multiplication (A * B):\n", np.dot(A, B))

# 4. Trace
print("\n4. Trace of A:", np.trace(A))

# 5. Hermitian Check
is_hermitian = np.allclose(A, np.conj(A.T))
print("\n5. Is Matrix A Hermitian?:", is_hermitian)

# 6. Unitary Check
identity = np.eye(A.shape[0])
is_unitary = np.allclose(np.dot(A, np.conj(A.T)), identity)
print("6. Is Matrix A Unitary?:", is_unitary)

# 7. Tensor Product (Kronecker Product)
print("\n7. Tensor Product (A ⊗ B):\n", np.kron(A, B))
```

**Task 2 : Get two matrix user and perform actions on thosse matrixes**

```python
import numpy as np

def get_square_matrix_input(size, name):
    print(f"\nEnter elements for Matrix {name} ({size}x{size}) row by row.")
    print("Separate elements within a row using spaces (e.g., '1+2j' or '4').")
    matrix = []
    for i in range(size):
        while True:
            try:
                line = input(f"Row {i+1}: ").strip().split()
                if len(line) != size:
                    print(f"Error: Expected {size} elements, but got {len(line)}. Try again.")
                    continue
                matrix.append([complex(x) for x in line])
                break
            except ValueError:
                print("Error: Invalid format. Enter again")
    return np.array(matrix)

print("Operations on given matrix")
n = int(input("Enter the size N for the square matrices (N x N): "))

A = get_square_matrix_input(n, "A")
B = get_square_matrix_input(n, "B")

print("\n" + "="*40)
print("1. MATRIX MULTIPLICATION")
print("="*40)
AB = np.dot(A, B)
BA = np.dot(B, A)
print("AB:\n", AB)
print("BA:\n", BA)
print("Are AB and BA equal?:", np.allclose(AB, BA))

print("\n" + "="*40)
print("2. TRANSPOSE & ADJOINT OPERATIONS")
print("="*40)
print("Transpose of A:\n", np.transpose(A))
print("Conjugate Transpose (Adjoint) of A:\n", np.conj(A.T))

print("\n" + "="*40)
print("3. TRACE, HERMITIAN & UNITARY CHECKS")
print("="*40)
print("Trace of A:", np.trace(A))
print("Is Matrix A Hermitian?:", np.allclose(A, np.conj(A.T)))
print("Is Matrix A Unitary?:", np.allclose(np.dot(A, np.conj(A.T)), np.eye(n)))

print("\n" + "="*40)
print("4. VECTOR-STYLE OPERATIONS (Flattened Matrices)")
print("="*40)
# Euclidean distance works directly on matrices of the same shape
print("Euclidean Distance between A and B:", np.linalg.norm(A - B))
# Orthogonality requires flattening the matrices to treat them like vectors
matrix_dot = np.vdot(A.flatten(), B.flatten())
print(f"Inner Product (A · B): {matrix_dot}")
print("Are they orthogonal?:", np.isclose(matrix_dot, 0))

print("\n" + "="*40)
print("5. TENSOR PRODUCT")
print("="*40)
print("Tensor product (A ⊗ B):\n", np.kron(A, B))
```

**Task 3 : Get one matrix input and find if it Hermitian or not.**

```python
import numpy as np

print("===  is Hermitian Matrix ? ===")
size = int(input("Enter the size of the square matrix (N x N): "))

print(f"Enter the elements row by row. Separate elements with spaces (e.g., 2-3j):")
matrix = []
for i in range(size):
    while True:
        try:
            line = input(f"Row {i+1}: ").strip().split()
            if len(line) != size:
                print(f"Error: Expected {size} elements, got {len(line)}. Try again.")
                continue
            matrix.append([complex(x) for x in line])
            break
        except ValueError:
            print("Error: Invalid input. Ensure numbers match formats like '4' or '2-3j'.")

M = np.array(matrix)
print("\nYour Matrix M:\n", M)

# Check condition: M == conj(M.T)
is_hermitian = np.allclose(M, np.conj(M.T))

if is_hermitian:
    print("\nResult: The matrix IS Hermitian.")
else:
    print("\nResult: The matrix IS NOT Hermitian.")
```

**Task 3 : Get one matrix input and find if it Unitary or not.**

```python
import numpy as np

print("=== Task 4: is Unitary Matrix ? ===")
size = int(input("Enter the size of the square matrix (N x N): "))

print(f"Enter the elements row by row. Separate elements with spaces (e.g., 0+0.5j):")
matrix = []
for i in range(size):
    while True:
        try:
            line = input(f"Row {i+1}: ").strip().split()
            if len(line) != size:
                print(f"Error: Expected {size} elements, got {len(line)}. Try again.")
                continue
            matrix.append([complex(x) for x in line])
            break
        except ValueError:
            print("Error: Invalid input format.")

M = np.array(matrix)
print("\nYour Matrix M:\n", M)

# Compute M * M^H
M_H = np.conj(M.T)
product = np.dot(M, M_H)
identity = np.eye(size)

is_unitary = np.allclose(product, identity)

print("\nProduct of M and its Conjugate Transpose (M * M^H):\n", product)

if is_unitary:
    print("\nResult: The matrix IS Unitary.")
else:
    print("\nResult: The matrix IS NOT Unitary.")
```
