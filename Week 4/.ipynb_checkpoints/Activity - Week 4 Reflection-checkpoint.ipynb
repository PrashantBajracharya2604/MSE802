```python
# Task : Create a 2-qubit quantum circuit with a 90° phase shift and visualize it using the Q-sphere.

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_state_qsphere

# 2-qubit circuit
qc_phase = QuantumCircuit(2)
qc_phase.h(0)  # Superposition
qc_phase.s(0)  # 90-degree phase shift

# Q-sphere
state = Statevector.from_instruction(qc_phase)
plot_state_qsphere(state)
```

```python
#Task : Create a circuit in qasm and import and run it in qiskit.

from qiskit import QuantumCircuit

#circuit in qasm string format
qasm_string = """
OPENQASM 2.0;
include "qelib1.inc";
qreg q[2];
h q[0];
cx q[0], q[1];
"""

# Import and run in qiskit
qc_from_qasm = QuantumCircuit.from_qasm_str(qasm_string)
print("Imported Circuit:")
display(qc_from_qasm.draw(output='mpl'))
```

## Reflection on Density Matrix – Diagonal and Off-Diagonal Elements.
### Density Matrix
The Bloch Sphere is the map of where a qubit could go, the Density Matrix is the diagnostic report of what the qubit is actually doing.
### Diagonal Elements
It represent the probability of the qubit being in state $|0\rangle$ or $|1\rangle$.
### Off-Diagonal Elements
It describes the interation of the states. It represents coherence and holds the relative phase. This is the information that tells us where the qubit is pointing along the equator.

```python
# Draw any circuit random. But don't use dump method.

from qiskit import QuantumCircuit
import numpy as np

# a random circuit
qc = QuantumCircuit(2)
qc.h(0)           # Superposition
qc.t(0)           # 45-degree rotation
qc.cx(0, 1)        # Entangle

# using Matplotlib
display(qc.draw(output='mpl'))
```
