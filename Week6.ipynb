```python
# Mannual ccx , swap,   hardcorde script


import numpy as np

## 1. MANUALLY HARDCODED 4x4 SWAP MATRIX (2 Qubits)
## Maps: |00> -> |00>, |01> -> |10>, |10> -> |01>, |11> -> |11>

manual_swap = np.array([
    [1, 0, 0, 0],  # Row for |00>
    [0, 0, 1, 0],  # Row for |01> -> Swaps with |10>
    [0, 1, 0, 0],  # Row for |10> -> Swaps with |01>
    [0, 0, 0, 1]   # Row for |11>
])

print("--- Manual SWAP Matrix (4x4) ---")
print(manual_swap)

## 2. MANUALLY HARDCODED 8x8 CCX (TOFFOLI) MATRIX (3 Qubits)
## Only flips the last two states: |110> <-> |111Object

manual_ccx = np.array([
    [1, 0, 0, 0, 0, 0, 0, 0],  # |000> stays same
    [0, 1, 0, 0, 0, 0, 0, 0],  # |001> stays same
    [0, 0, 1, 0, 0, 0, 0, 0],  # |010> stays same
    [0, 0, 0, 1, 0, 0, 0, 0],  # |011> stays same
    [0, 0, 0, 0, 1, 0, 0, 0],  # |100> stays same
    [0, 0, 0, 0, 0, 1, 0, 0],  # |101> stays same
    [0, 0, 0, 0, 0, 0, 0, 1],  # |110> flips to |111>
    [0, 0, 0, 0, 0, 0, 1, 0]   # |111> flips to |110>
])

print("\n--- Manual CCX Matrix (8x8) ---")
print(manual_ccx)

## EXAMPLE 1: SWAP

initial_state_2q = np.array([0, 1, 0, 0])
final_state_2q = manual_swap @ initial_state_2q
print("SWAP\n")
print("Initial statevector:", initial_state_2q)
print("Final statevector :", final_state_2q)
print("Result: swapped successfully!\n")

## EXAMPLE 2: CCX (Toffoli)
initial_state_3q = np.array([0, 0, 0, 0, 0, 0, 1, 0])
final_state_3q = manual_ccx @ initial_state_3q
print("CCX\n")
print("Initial statevector:", initial_state_3q)
print("Final statevector:  ", final_state_3q)
```
